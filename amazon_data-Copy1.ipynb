```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Read vaccinations and government_response CSV files into DataFrames
vaccinations_df = pd.read_csv("C:/Users/leahm/Downloads/vaccinations.csv")
government_response_df = pd.read_csv("C:/Users/leahm/Downloads/oxford-government-response.csv")

# Merge the two DataFrames based on the 'Country' column
merged_df = pd.merge(vaccinations_df, government_response_df, on='location_key')

# Assuming your merged DataFrame has columns 'Country', 'Latitude', 'Longitude', and other relevant columns
# You may need to adjust the column names based on your actual data

# Create a GeoDataFrame from the merged DataFrame
gdf = gpd.GeoDataFrame(merged_df, 
                       geometry=gpd.points_from_xy(merged_df['Longitude'], merged_df['Latitude']))

# Plot the map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
fig, ax = plt.subplots(figsize=(15, 10))

# Plot world map
world.plot(ax=ax, color='lightgrey')

# Plot the points on the map
gdf.plot(ax=ax, marker='o', color='red', markersize=10, label='Vaccination and Response Data')

# Add legend
ax.legend()

# Add title and labels
plt.title('Vaccination and Government Response Data Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the map
plt.show()
```

```python
genres_df = amazon_df[["genres1", "genres2", "genres3", "genres4", "genres5", "genres6", "genres7", "genres8", "genres9"]]
genres_df = genres_df.fillna('')
genres_df.head()
```

```python
new_genres_df = genres_df.stack().reset_index()
new_genres_df.columns = ["movie_number", "column", "genre"]
new_genres_df["genre"] = new_genres_df["genre"].str.strip()
sorted_df = new_genres_df.loc[(new_genres_df["genre"] != "")]
new_sorted_df = sorted_df.groupby('genre')
```
